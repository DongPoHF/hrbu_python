```python
path='./U'
#
new_path='./U/'
for root ,dirs,files in os.walk(path):
    for i in range(len(files)):
        if (files[i][-3:] == 'jpg'):
            file_path = root +'/'+files[i]
            #print(file_path)
            if not os.path.exists(new_path):
                os.mkdir(new_path)
                shutil.move(file_path,new_path)
                #new_file_path = new_path+'/'+files[i]
                #print(new_file_path)
            else:
                shutil.move(file_path,new_path)
                
#
new_path2='./U/'
for root ,dirs,files in os.walk(path):
    for i in range(len(files)):
        if (files[i][-3:] == 'doc'):
            file_path = root +'/'+files[i]
            #print(file_path)
            if not os.path.exists(new_path2):
                os.mkdir(new_path2)
                shutil.move(file_path,new_path2)
            else:
                shutil.move(file_path,new_path2)

#
new_path3='./U/'
for root ,dirs,files in os.walk(path):
    for i in range(len(files)):
        if (files[i][-3:] == 'xls'):
            file_path = root +'/'+files[i]
            #print(file_path)
            if not os.path.exists(new_path3):
                os.mkdir(new_path3)
                shutil.move(file_path,new_path3)
            else:
                shutil.move(file_path,new_path3)

#
path2='./U./_opop/test -  (2)/link'
new_path4='./U/'
list=os.listdir(path2)
for i in list:
    if(i[-3:]==i.split('.')[-1]):
        file_path = path2 +'/'+i
        #print(file_path)
        if not os.path.exists(new_path4):
            os.mkdir(new_path4)
            shutil.move(file_path,new_path4)
        else:
            shutil.move(file_path,new_path4)


#
for root,dirs,files in os.walk(path):
    for file in files:
        src_file=os.path.join(root,file)
        if os.path.getsize(src_file) == 0:
            os.remove(src_file)
dir_list=[]
for root,dirs,files in os.walk(path):
    dir_list.append(root)
    #print(dir_list)
# 
for root in dir_list[::-1]:
    if not os.listdir(root):
        os.rmdir(root)
```
