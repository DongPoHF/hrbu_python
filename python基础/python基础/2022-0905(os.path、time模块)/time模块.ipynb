```python
import time
```

```python
#时间模块的值
#timezone
# 获取当前时区于格林尼治时区相差的
time.timezone
```

```python
#daylight  夏令时
time.daylight
```

```python
#时间模块的函数
#asctime()
res=time.localtime()
time.asctime(res)
```

```python
# localtime()
#获取当前时间元组
#格式1：获取本地时间元组
time.localtime()
#格式2：获取指定的时间元组
#time.localtime(时间戳) 
```

```python
# gmtime
# 获取当前utc时间的时间元组
#格式1：获取当前utc时间的时间元组
time.gmtime()
#格式2：获取指定的时间utc元组
#time.gmtime(时间戳)
```

```python
#ctime()
#获取本地时间的字符串
time.ctime()
```

```python
# 利用时间元组制作时间戳
time.mktime(time.localtime())
```

```python
#sleep() 秒
#程序睡眠，使得程序在此处等待的时间
#time.sleep(2)
for i in range(0,10):
    time.sleep(2)
    print(i)
```
